### Evidence Statistics 

```python
# load all evidence files
import os
import pandas as pd

parent_folder = os.path.join('..', 'src', 'cli', 'outputs')
last_folder = max([f for f in os.listdir(parent_folder) 
                   if os.path.isdir(os.path.join(parent_folder, f))])
last_folder
```

```python
last_run = os.path.join(parent_folder, last_folder)
# Flight - Message - Sentence - Unique Sentence - Problems - Unique Problems.
evidence_dirs = [f for f in os.listdir(last_run)]
rows = []
for evidence_dir in evidence_dirs:
    evidence_path = os.path.join(last_run, evidence_dir, 'final.xlsx')
    evidence_df = pd.read_excel(evidence_path)
    problem_df = evidence_df[evidence_df['problem_type'] != 'Normal']
    unique_problems = problem_df['sentence'].unique()
    unique_problem_types = problem_df['problem_type'].unique()
    rows.append({
        'flight': evidence_dir,
        'message': evidence_df['message_id'].nunique(),
        'sentence': evidence_df['sentence'].count(),
        'problems': problem_df['sentence'].count(),
        'unique_sentences': evidence_df['sentence'].nunique(),
        'unique_problems': problem_df['sentence'].nunique(),
        'key_events': unique_problems.tolist(),
        'problem_types': unique_problem_types.tolist(),
    })
summary_df = pd.DataFrame(rows, columns=['flight', 'message', 'sentence', 'problems', 'unique_sentences', 'unique_problems', 'key_events', 'problem_types'])
# summary_df = summary_df.sort_values(by='flight', ascending=False)
```

```python
summary_df.to_excel(os.path.join('..', 'experiments', 'analysis', 'evidence_summary.xlsx'), index=False)
```
